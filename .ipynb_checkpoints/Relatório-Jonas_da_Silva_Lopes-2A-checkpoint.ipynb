## Introdução

A sociedade é um objeto muito complexo, por tal motivo existe uma ciência destinada somente a seu estudo e compreensão. Nessa ciêcia um de seus maiores temas é o estudo do poder da pressão social feita sobre um índivíduo, ou de um grupo sobre outro, destacando ainda o poder de construção da essência de um ser por meio da sua socialização. Assim, atualmente pode ser discutido os diversos tipos de pressões socias, sejam elas causadoras de efeitos positivo e negativo, e dentre as consequências negativas há a possibilidade de suicídio.
Dentre essas pressões, pode-se citar como uma das mais famosas e estudadas, a pressão da família pra que um jovem consiga ser bem sucedido profissionalmente, que pode ser vista nos artigos "Stress and suicidal ideation among adolescents having academic difficulty"(https://www.ncbi.nlm.nih.gov/pmc/articles/PMC5810170), "Parental Pressure on Students"(https://www.pewresearch.org/global/2006/08/24/parental-pressure-on-students), "Youth suicide: I tried to kill myself three times by 17"(https://www.bbc.com/news/uk-46812837). Todas essas referências tem como principais causas do suicídio, o desenvolvimento profissional, especialmente em países economicamente desenvolvidos, como no caso dos países integrantes do G7.

## Pergunta a ser trabalhada:

Os países integrantes do G7 tendem a possuir uma maior taxa de suicídio quando comparados a países que não são integrantes?

## Mecanismo

Os países que são mais desenvolvidos, como no caso dos países que formam o G7, podem tender a exigirem demandas altas de sucesso pessoal, seja: profissional, interpessoal, acadêmico, financeiro, entre outros. Essa coerção em uma nação tem o potencial de criar uma grande pressão psicológica nas pessoas, o que as faz pensar que seu único objetivo de vida seja alcançar o sucesso nessas áreas da vida, ou que para se ter uma boa vida é necessário atingir esses "critérios", assim por muitas vezes pessoas que "falham" com suas metas, ou pensam que essas são inalcançáveis tiram suas próprias vidas, tendo em vista que seus únicos objetivos na vida não podem ser alcançados. Logo, pode-se afirmar que o desenvolvimento econõmico dos países que estão no G7 aumenta a incidência de suicídios neles devido a uma grande pressão social promovida pela sociedade.

## Comparando o Pib de um país e seu Indice de suícidios

### Análise dos gráficos de dispersão com reta de ajuste e coeficiente de correlação

#### Função para plotagem dos gráficos

```python
def reta(dados_x, dados_y, c):    
    a = dados_y.cov(dados_x) / dados_x.var()
    b = dados_y.mean() - a*dados_x.mean()

    print('Ajuste linear entre X3 e X4:')
    print('Coeficiente linear:', b)
    print('Coeficiente angular:', a)
    plt.scatter(dados_x, dados_y, c=c, alpha=0.9)
    plt.title('x versus y')
    plt.xlabel('eixo x')
    plt.ylabel('eixo y')

    plt.plot((dados_x.min(), dados_x.max()), (a*dados_x.min()+b, a*dados_x.max()+b), color=c)

    plt.tight_layout()
```

#### Análise do ano 1998

```python
reta( dados_gerais.loc[:,'ano1998PIB'],dados_gerais.loc[:,'ano1998Suicídios'], "r")
plt.title('X versus Y: 1998')
plt.xlabel('PIB per capita')
plt.ylabel('Suicídio por cem mil pessoas')
coeficiente_de_correlação = dados_gerais.loc[:,'ano1998PIB'].corr(dados_gerais.loc[:,'ano1998Suicídios'])
print(('O coeeficiente de correlação é: {0:.4f}').format(coeficiente_de_correlação))
```

#### Análise do ano 2000

```python
reta( dados_gerais.loc[:,'ano2000PIB'], dados_gerais.loc[:,'ano2000Suicídios'], "black")
plt.title('X versus Y: 2000')
plt.xlabel('PIB per capita')
plt.ylabel('Suicídio por cem mil pessoas')
coeficiente_de_correlação = dados_gerais.loc[:,'ano2000PIB'].corr(dados_gerais.loc[:,'ano2000Suicídios'])
print(('O coeeficiente de correlação é: {0}').format(coeficiente_de_correlação))
```

## Gráfico stacked

O gráfico do tipo stacked é uma excelente ferramenta para fazer uma análise relativa entre as taxas de suícidio do G7 e as taxas de suicídios dos países não integrantes do G7

#### Primeiramente deve-se criar dataframes com as médias de cada ano dos países do G7 e dos não pertencentes ao G7

```python
sui_medias_g7 = sui_limpg7.mean()
sui_medias_fora = sui_limpfora.mean()
medias = {'ano1998':[sui_medias_g7[0], sui_medias_fora[0]], 'ano2000': [sui_medias_g7[1], sui_medias_fora[1]], 'ano2002':[sui_medias_g7[2], sui_medias_fora[2]], 'ano2004':[sui_medias_g7[3], sui_medias_fora[3]]}                        
data_medias = pd.DataFrame(medias, columns=['ano1998','ano2000','ano2002','ano2004'])
```

#### Após esse processo deve-se normalizar os dados

```python
somas = data_medias.iloc[:,0:4].sum()
medias2 = (data_medias/somas)*100
```

#### Criar uma coluna para categorizar os países entre participantes e não participantes do G7

```python
medias2['Tipo'] = ['G7', 'não pertence ao G7']
```

#### Definir a coluna "Tipo" como índice, para facilitar a visualização na plotagem do gráfico

```python
new2 = medias2.set_index('Tipo')
```

#### Plotagem do gráfico stacked

```python
dados_gerais_stacked = new2
plot = dados_gerais_stacked.T.plot(kind='bar', stacked=True,title='Porcentagem de suícidos no ano por categoria "tipo"', figsize=(6,6))
plot.set_ylim([0, 100])
plt.ylabel('Freq.Relativa (em%)')

#Para legenda ficar fora do gráfico
ax = plt.subplot(111)
box = ax.get_position()
ax.set_position([box.x0, box.y0, box.width * 0.8, box.height])
ax.legend(loc='center left', bbox_to_anchor=(1, 0.5))
```

O gráfico stacked é bem claro com relação a comparação de suicídios, sendo que em todos os anos analisados o grupo dos G7 representam aproximadamente 60% dos suicídios totais analisados.
